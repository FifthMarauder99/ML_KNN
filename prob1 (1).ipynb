Step 2: Choosing Random points

```python
def randomCentroid(dataset, k):
    #Initially created empty centroid
    centroids = []
    for i in range(k):
        centroid = dataset.apply(lambda f: float(f.sample()))
        print(centroid) 
        #appending empty centroid after calulating
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
```

Now, calculating the distance using distance formula

```python
# referred from geeks for geeks,for implementing the formula inside lambda 
import numpy as np
def label(dataset, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((dataset - x) ** 2).sum(axis=1)))
    #idxmin is used to return series of with the index of the minimum value for each column
    return distances.idxmin(axis=1)
```

```python
labels = label(dataset, centroids)
```

step 3: Recompute the centroids of newly formed clusters

```python
#group by is numpy function to cluster them
#https://www.analyticsvidhya.com/blog/2019/08/comprehensive-guide-k-means-clustering/
def NewCentroids(dataset, labels, k):
    centroids = dataset.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
    return centroids
```

```python
def plotCluster(dataset, labels, centroids, iteration):
    pca = PCA(n_components=2)
    # transform data to plot it
    dataIn2d = pca.fit_transform(dataset)
    # transform centroid and transposing
    centroidin2d = pca.transform(centroids.T)
    #to show desired output and clearning all the unwanted graphs
    clear_output(wait=True)
    plot.title(f'Iteration {iteration}')
    plot.scatter(x=dataIn2d[:,0], y=dataIn2d[:,1], c=labels)
    plot.scatter(x=centroidin2d[:,0], y=centroidin2d[:,1])
    plot.show()
#     print(data_2d)
#     print(centroids_2d)
```
